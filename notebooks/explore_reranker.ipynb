# Explore Reranker

This notebook inspects reranker scores and latency on a small local sample.

```python
import json
from pathlib import Path

from reranker.inference import ONNXReranker
```

```python
corpus_path = Path("../data/corpus.json")
if not corpus_path.exists():
    raise FileNotFoundError("Run moss_indexer first to create data/corpus.json")

docs = json.loads(corpus_path.read_text(encoding="utf-8"))[:8]
query = "How do I create an index in Moss?"
reranker = ONNXReranker("../reranker/models/reranker_onnx_quantized")
top_docs, latency_ms = reranker.rerank(query, docs, top_k=3)
latency_ms
```

```python
for i, doc in enumerate(top_docs, start=1):
    print(f"{i}. id={doc.get('id')} score={doc.get('rerank_score'):.4f}")
    print((doc.get('text') or '')[:220])
    print()
```
